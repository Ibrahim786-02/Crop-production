```python
import numpy as np
import pandas as pd
data=pd.read_csv("Crop Production data.csv")
print(data.head())
```

```python
summary_stats = data.describe()
summary_stats
```

```python

missing_data = data.isnull().sum()
missing_data
```

```python
import matplotlib.pyplot as plt

# Group by Crop_Year and sum the Production for each year
yearly_production = data.groupby('Crop_Year')['Production'].sum()

# Plotting the trend of crop production over the years
plt.figure(figsize=(10, 6))
plt.plot(yearly_production.index, yearly_production.values, marker='o', color='blue')
plt.title('Total Crop Production Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Production (Tons)')
plt.grid(True)
plt.show()

yearly_production
```

```python
# Group by Season and sum the Production for each season
seasonal_production = data.groupby('Season')['Production'].sum()

# Plotting the seasonal crop production
plt.figure(figsize=(10, 6))
seasonal_production.plot(kind='bar', color='blue')
plt.title('Total Crop Production by Season')
plt.xlabel('Season')
plt.ylabel('Total Production (Tons)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Display the seasonal production values
print(seasonal_production)
```

```python
# Group by State_Name and sum the Production for each state
statewise_production = data.groupby('State_Name')['Production'].sum().sort_values(ascending=False)

# Plotting the state-wise crop production
plt.figure(figsize=(15, 8))
statewise_production.plot(kind='bar', color='green')
plt.title('Total Crop Production by State')
plt.xlabel('State')
plt.ylabel('Total Production (Tons)')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()

# Display the state-wise production values
print(statewise_production)
```
